# Boolean indexing
1. ðŸ”¹ Boolean Indexing kya hota hai?

**Jab hum NumPy array ya Pandas DataFrame ke elements ko True/False condition ke basis par filter karte hain, usse Boolean Indexing bolte hain.
Matlab hum ek condition lagayenge aur jo condition True hogi wahi values return hongi.**


```python
import numpy as np 
a=np.random.randint(1,100,24).reshape(6,4) # randint(range h 1, se 100 ki , fir kitne no. chahiye )
a
```

```python
#find all number greater than 50
a>50
```

```python
#ab hum is consition ko indexing ki tarah use me le skate hai 
a[a>50]
```

```python
#find out even numbers 
a[a%2==0]
```

```python
#find the numbers greater than 50 and even 
a[(a>50) & (a%2==0)] #jab bhi boolean values ki baat hogi to '&' use karenge 'and' nahi 
```

```python
#Not divisible by 7 or 50 se bada 
a[~(a%7==0)&(a>50)]
```

# Broadcasting
ðŸ”¹ Broadcasting Kya Hai?

Broadcasting ek NumPy feature hai jo alag-alag size ke arrays ke beech operations ko allow karta hai without explicit looping.
Matlab agar arrays ka shape alag hai, to NumPy automatically expand/adjust karke operation kar deta hai.

```python
#same shape
a=np.arange(6).reshape(2,3)
b=np.arange(6,12).reshape(2,3)

print(a)
print(b)
print(a+b) #itemwise edition
```

```python
#different shape
a=np.arange(6).reshape(2,3)
b=np.arange(3).reshape(1,3)

print(a)
print(b)
print(a+b)
```

# ðŸ”¹ Broadcasting Rule

NumPy arrays ke shapes ko right se compare karta hai:
 1.	Agar dimensions same hain â†’ operation ho jayega.
 2.	Agar ek dimension me difference hai aur ek side me 1 hai â†’ wo dimension stretch ho jayega.
 3.	Agar mismatch hai (aur kahin 1 nahi hai) â†’ error milega.

```python
#More examples 
a=np.arange(12).reshape(4,3)
b=np.arange(3).reshape(1,3)
print(a)
print(b)
print(a+b)
```

```python
#changing shape  
a=np.arange(12).reshape(3,4)
b=np.arange(3)
print(a+b)
```

```python
#mean Square error
actual=np.random.randint(1,50,25)
predicted=np.random.randint(1,50,25)
actual
```

```python
predicted
```

```python
def mse(actual,predicted):
    return np.mean((actual-predicted)**2)

mse(actual,predicted)
```

```python
#binary cross Entropy

def bce(actual,predicted):
  return  -np.mean(actual*np.log(predicted)+(1-actual)*np.log(1-predicted))

actual=np.array([0,1,0,1,1,0])
predicted=np.array([0.9,0.1,0.3,0.3,0.8,0.6])

bce(actual,predicted)
```
