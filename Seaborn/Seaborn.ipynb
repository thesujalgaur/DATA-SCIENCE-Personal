# Relational Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
#scatter plot
tips=sns.load_dataset('tips')
tips
```

```python
sns.scatterplot(tips,x='total_bill',y='tip',hue='sex',style='time')
```

```python
sns.relplot(tips,x='total_bill',y='tip',kind='scatter',hue='sex',style='day',size='size')
```

# Line plot

```python
gap=px.data.gapminder()
gap
```

```python
df=gap[gap['country']=='India']
df
```

```python
#axis level function
sns.lineplot(df,x='year',y='lifeExp')
```

```python
sns.relplot(df,x='year',y='lifeExp',kind='line')
```

```python
df=gap[gap['country'].isin(['India','Brazil','Germany','China'])]
df
```

```python
sns.relplot(df,x='year',y='lifeExp',kind='line',hue='country',style='continent',size='continent')
```

```python
sns.lineplot(df,x='year',y='lifeExp',hue='country')
```

# matrix plot

```python
#Heatmap
```

```python
gap
```

```python
temp_df=gap.pivot(index='country',columns='year',values='lifeExp')
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(temp_df)
```

```python
AS=gap[gap['continent']=='Europe'].pivot(index='country',columns='year',values='lifeExp')
```

```python
plt.figure(figsize=(15,7))
sns.heatmap(AS,annot=True,linewidths=2,linecolor='black',cmap='winter')
```
