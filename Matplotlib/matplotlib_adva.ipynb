```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris=pd.read_csv('iris.csv')
iris
```

```python
iris['Species']=iris['Species'].replace({'Iris-setosa':0,
                         'Iris-versicolor':1,
                         'Iris-virginica':2})
iris.sample(5)
```

```python
plt.figure(figsize=(15,7))
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'],cmap='twilight',alpha=0.9)
plt.xlabel('sepal Length')
plt.ylabel('Petal Length')
plt.colorbar()
#alpha is opacity ( Density)
```

# Annotations
*isko use karke hum apne graph pe ek text ko place kar skte h*

```python
batter=pd.read_csv('batter.csv')
batter
```

```python
sample_df=batter.head(100).sample(25,random_state=5)
```

```python
sample_df
```

```python
plt.figure(figsize=(18,5))
plt.scatter(sample_df['avg'],sample_df['strike_rate'],s=sample_df['runs'])
for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])
```

```python
x=[1,2,3,4]
y=[5,6,7,8]
plt.scatter(x,y)
plt.text(1,5,'Point 1')
plt.text(2,6,'Point 2')
plt.text(3,7,'Point 3')
plt.text(4,8,'Point 4',fontdict={'size':14,'color':'Red'})
```

```python
#horizontal and vertical line
plt.figure(figsize=(18,5))
plt.scatter(sample_df['avg'],sample_df['strike_rate'],s=sample_df['runs'])
for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])
plt.axhline(130,color='red')
plt.axhline(140,color='black')
plt.axvline(30,color='red')
```

```python

fig,ax=plt.subplots(nrows=2,ncols=2,figsize=(10,10))
ax[0,0].scatter(batter['avg'],batter['strike_rate'],color='brown',marker='D')
ax[0,1].scatter(batter['avg'],batter['runs'],color='pink')
ax[1,0].hist(batter['avg'],color='pink')
ax[1,1].hist(batter['runs'],color='pink')
ax[0].set_title('avg vs strike rate')
ax[1].set_title('avg vs runs')
ax[0].set_xlabel('Strike rate')
ax[0].set_ylabel('avg run')
ax[1].set_xlabel('avg')
ax[1].set_ylabel('runs')
```
