```python
batsman=pd.read_csv('sharma-kohli.csv')
batsman
```

```python
#plotting multiple plots
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
```

```python
#lables title
plt.plot(batsman['index'],batsman['V Kohli'],color='green',label='NUmber 18')
plt.plot(batsman['index'],batsman['RG Sharma'],color='orange',label='NUmber 45')
plt.title('hum rohit sharma aur virat kohli ka career comparison kar rhe h ')
plt.xlabel('season')
plt.ylabel('total run')
plt.legend()
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='green',linestyle='solid',linewidth=3)
plt.plot(batsman['index'],batsman['RG Sharma'],color='blue',linestyle='dashdot',linewidth=2)
plt.title('hum rohit sharma aur virat kohli ka career comparison kar rhe h ')
plt.xlabel('season')
plt.ylabel('total run')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='green',linestyle='solid',linewidth=3,marker='D',markersize=10)
plt.plot(batsman['index'],batsman['RG Sharma'],color='blue',linestyle='dashdot',linewidth=2,marker='o',markersize=9)
plt.title('hum rohit sharma aur virat kohli ka career comparison kar rhe h ')
plt.xlabel('season')
plt.ylabel('total run')
```

```python
#legend --> location
plt.plot(batsman['index'],batsman['V Kohli'],color='green',linestyle='solid',linewidth=3,marker='D',markersize=10,label='virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color='blue',linestyle='dashdot',linewidth=2,marker='o',markersize=9,label='Rohit')
plt.title('hum rohit sharma aur virat kohli ka career comparison kar rhe h ')
plt.xlabel('season')
plt.ylabel('total run')
plt.legend(loc='lower right')
```

# Bar chart

```python
#simple bar chart
children=[10,20,12,32,45]
colors=['red','blue','green','yellow','pink']
plt.bar(colors,children,color=colors)
plt.title('konse bache ko konsa color pasand h ')
```

```python
#horizontal bar chart 
#simple bar chart
children=[10,20,12,32,45]
colors=['red','blue','green','yellow','violet']
plt.barh(colors,children,color=colors)
plt.title('konse bache ko konsa color pasand h ')
```

```python
plt.bar(np.arange(df.shape[0])-0.2,df['2015'],width=0.2,color="yellow")
plt.bar(np.arange(df.shape[0]),df['2016'],width=0.2,color='pink')
plt.bar(np.arange(df.shape[0])+0.2,df['2017'],width=0.2,color='violet')

plt.xticks(np.arange(df.shape[0]),df['batsman'])
plt.show()
```

```python
#stacked bar chart 


plt.bar(df['batsman'],df['2017'],label='2017')
plt.bar(df['batsman'],df['2016'],bottom=df['2017'],label='2016')
plt.bar(df['batsman'],df['2015'],bottom=df['2016']+df['2017'],label='2015')

plt.legend()
plt.show()
```
