# Working with multiindexed dataframe

```python



branches = ["cse", "ece"]
years = [2019, 2020, 2021, 2022]
index = pd.MultiIndex.from_product([branches, years], names=["branch", "year"])


cities = ["delhi", "mumbai"]
metrics = ["avg_package", "students"]
columns = pd.MultiIndex.from_product([cities,metrics], names=["city", "metric"])


data = [
    [1, 2, 0, 0],
    [3, 4, 0, 0],
    [5, 6, 0, 0],
    [7, 8, 0, 0],
    [9, 10, 0, 0],
    [11, 12, 0, 0],
    [13, 14, 0, 0],
    [15, 16, 0, 0],
]

df = pd.DataFrame(data, index=index, columns=columns)
print(df)
```

# Long V/s Wide Data
1. ***Wide format is where we have a single row for  every data point with multiple columns to hold the values of various attributes.***
2. **Long format is where, for each data point we have as many rows as the number of attributes and each row contains the value of a particular attribute for a given data point.**

```python
#MELT--> Wide data format ko long me convert karta h 
pd.DataFrame({'cse':[20]}).melt()
```

```python
pd.DataFrame({
    'cse':[120],
    'ece':[234],
    'me':[22]
}).melt(var_name='branch',value_name='students')
```

```python
pd.DataFrame({
    'Branch':['Ece','Bce','mec'],
    '2020':[11,2,3],
    '2021':[23,32,33],
    '2025':[45,96,232]
    }).melt(id_vars=['Branch'],var_name='year',value_name='students')
```
